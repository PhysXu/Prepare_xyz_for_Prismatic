```python
import pandas as pd
```

```python
column_names = ['Element', 'X', 'Y', 'Z']
coords  = pd.read_excel(r'C:\Users\STEM\Desktop\LiCoO2 initial.xlsx', header=None, names=column_names)
coords
```

```python
# Add the occupancy and isotropic thermal parameters

coords['occupancy']=1
coords['isotropic thermal parameters']=0.08
coords
```

```python
# If you want use different Uiso for different elements, run the Cell

coords.loc[coords['Element'] == 'Li', 'isotropic thermal parameters'] = 0.06
coords
```

```python
# Change the element to the element_No.

coords['Element'] = coords['Element'].replace({'Li': 3, 'Co': 27, 'O': 8})
coords
```

```python
fileName = r'C:\Users\STEM\Desktop\LiCoO2 final.xyz'
comment = 'LiCoO2'
cellDim = [14.214, 24.620, 14.146]   # the size of cell

with open(fileName, 'w') as fid:
    # Write comment line (1st)
    fid.write(f"{comment}\n")

    # Write cell dimensions
    fid.write(f"    {cellDim[0]} {cellDim[1]} {cellDim[2]}\n")

    # Write atomic data
    dataAll = coords.values
    for row in dataAll:
        fid.write(f"{int(row[0])}  {row[1]:.6f}  {row[2]:.6f}  {row[3]:.6f}  {int(row[4])}  {row[5]:.6f}\n")

    # Write end of file, for compatibility
    fid.write("-1\n")
```
